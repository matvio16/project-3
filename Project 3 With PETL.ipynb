```python
import petl as etl
import pandas as pd
import numpy as np
```

```python
#Read the CSV file
data = etl.fromcsv("countries of the world.csv")
```

```python
#Rename fields
data = etl.rename(data, 
{
    'Pop. Density (per sq. mi.)' : 'Pop Density',
    'Coastline (coast/area ratio)' : 'Coastline (ratio)',
    'Net migration' : 'Net Migration',
    'Infant mortality (per 1000 births)' : 'Infant Mortality (per 1000)',
    'GDP ($ per capita)' : 'GDP Per Capita'
})
data
```

```python
#Convert commas to periods
data = etl.convert(data, 'Pop Density', 'replace', ',', '.')
data = etl.convert(data, 'Coastline (ratio)', 'replace', ',', '.')
data = etl.convert(data, 'Net Migration', 'replace', ',', '.')
data = etl.convert(data, 'Infant Mortality (per 1000)', 'replace', ',', '.')
data = etl.convert(data, 'Literacy (%)', 'replace', ',', '.')
data = etl.convert(data, 'Phones (per 1000)', 'replace', ',', '.')
data = etl.convert(data, 'Arable (%)', 'replace', ',', '.')
data = etl.convert(data, 'Crops (%)', 'replace', ',', '.')
data = etl.convert(data, 'Other (%)', 'replace', ',', '.')
data = etl.convert(data, 'Climate', 'replace', ',', '.')
data = etl.convert(data, 'Birthrate', 'replace', ',', '.')
data = etl.convert(data, 'Deathrate', 'replace', ',', '.')
data = etl.convert(data, 'Agriculture', 'replace', ',', '.')
data = etl.convert(data, 'Industry', 'replace', ',', '.')
data = etl.convert(data, 'Service', 'replace', ',', '.')
data
```

```python
#Convert data types
data = etl.convert(data, 'Population', int, errorvalue=0)
data = etl.convert(data, 'Area (sq. mi.)', int, errorvalue=0)
data = etl.convert(data, 'Pop Density', float, errorvalue=0)
data = etl.convert(data, 'Coastline (ratio)', float, errorvalue=0)
data = etl.convert(data, 'Net Migration', float, errorvalue=0)
data = etl.convert(data, 'Infant Mortality (per 1000)', float, errorvalue=0)
data = etl.convert(data, 'GDP Per Capita', float, errorvalue=0)
data = etl.convert(data, 'Literacy (%)', float, errorvalue=0)
data = etl.convert(data, 'Phones (per 1000)', float, errorvalue=0)
data = etl.convert(data, 'Arable (%)', float, errorvalue=0)
data = etl.convert(data, 'Crops (%)', float, errorvalue=0)
data = etl.convert(data, 'Other (%)', float, errorvalue=0)
data = etl.convert(data, 'Climate', float, errorvalue=0)
data = etl.convert(data, 'Birthrate', float, errorvalue=0)
data = etl.convert(data, 'Deathrate', float, errorvalue=0)
data = etl.convert(data, 'Agriculture', float, errorvalue=0)
data = etl.convert(data, 'Industry', float, errorvalue=0)
data = etl.convert(data, 'Service', float, errorvalue=0)
data
```

```python
#Remove leading and trailing spaces in remaing strings
data = etl.convert(data, 'Country', lambda x : x.strip())
data = etl.convert(data, 'Region', lambda x : x.strip())
data
```

```python
etl.tocsv(data, "Cleaned Data.csv")
```

```python
df = pd.read_csv("Cleaned Data.csv")
df.head()
```

```python
countries = df['Country'].unique()
print(countries)
```

```python
print(len(countries))
```

```python
country_ids = np.arange(1, 228)
print(country_ids)
```

```python
c_ids = ['country' + str(id) for id in country_ids]
print(c_ids)
```

```python
country_df = pd.DataFrame({'country_id' : c_ids, 'country' : countries})
country_df
```

```python
regions = df['Region'].unique()
print(regions)
```

```python
print(len(regions))
```

```python
region_ids = np.arange(1, 12)
print(region_ids)
```

```python
r_ids = ['region' + str(id) for id in region_ids]
print(r_ids)
```

```python
region_df = pd.DataFrame({'region_id' : r_ids, 'region' : regions})
region_df
```

```python
#Export to csv
country_df.to_csv("Countries.csv", index=False)

region_df.to_csv("Regions.csv", index=False)
```

```python
c_df = country_df[['country', 'country_id']]
c_df = c_df.set_index('country')
c_df
```

```python
c_dict = c_df.to_dict()
c_dict
```

```python
df['Country ID'] = df['Country'].map(c_dict['country_id'])
df
```

```python
r_df = region_df[['region', 'region_id']]
r_df = r_df.set_index('region')
r_df
```

```python
r_dict = r_df.to_dict()
r_dict
```

```python
df['Region ID'] = df['Region'].map(r_dict['region_id'])
df
```
